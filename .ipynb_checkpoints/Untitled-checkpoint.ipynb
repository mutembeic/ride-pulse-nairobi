```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Train.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Placement - Time'] = pd.to_datetime(df['Placement - Time'])
df['Arrival at Destination - Time'] = pd.to_datetime(df['Arrival at Destination - Time'])
```

```python
df.head()
```

```python
# Convert day column to padded string (e.g., 9 -> 09)
df['Placement - Day of Month'] = df['Placement - Day of Month'].astype(str).str.zfill(2)

# Convert time to string (just in case it's datetime)
df['Placement - Time'] = df['Placement - Time'].astype(str)

# Combine into full datetime string
df['placement_datetime_str'] = '2019-03-' + df['Placement - Day of Month'] + ' ' + df['Placement - Time']

# Let pandas infer the format automatically
df['placement_datetime'] = pd.to_datetime(df['placement_datetime_str'], errors='coerce')
```

```python
df.drop(columns=['Temperature', 'Precipitation in millimeters'], inplace=True)
```

```python
rider_trip_counts = df.groupby('Rider Id').size().reset_index(name='Trip Count')
rider_trip_counts 
```

```python
rider_summary = df.groupby('Rider Id').agg({
    'Distance (KM)': 'mean',
    'Time from Pickup to Arrival': 'mean'
}).reset_index()
rider_summary
```
